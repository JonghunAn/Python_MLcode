```python
import matplotlib.pyplot as plt

%matplotlib inline

x_data =[x for x in range(-5,5)]
y_data =[y*y for y in range(-5,5)]

plt.title('line plot')
plt.grid()
plt.plot(x_data,y_data,color='b')
plt.show()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

x_data = [x for x in range(-4,5)]
y_data1 = [x*x for x in x_data]
y_data2 = [(x-1)*(x-1)+3 for x in x_data]

plt.title('line overwrite')
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xlim(-4,4)
plt.ylim(0,20)

plt.plot(x_data,y_data1,label='x*x')
plt.plot(x_data,y_data2,label='(x-1)*(x-1)+3')

plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

x_data = [x for x in np.arange(0,15,0.1)]
y1_data = [y1 for y1 in np.sin(x_data) ]
y2_data = [y2 for y2 in np.cos(x_data)]

plt.title('sin(x)/cos(x)')
plt.grid()
plt.xlabel('time')
plt.ylabel('amplitude')

plt.plot(x_data,y1_data,label='sin',linestyle='--',color='r')
plt.plot(x_data,y2_data,label='cos',linestyle='--',color='g')
plt.legend(loc='lower right')
plt.show()
```
